```python
import pandas as pd 
```

```python
# Load dataset
gp_poi = pd.read_csv('gp_pois_merged.csv')
gp_poi.head()
```

```python
nyc_geo = pd.read_csv('nyc_geo_latlon.csv') # neighbourhoods
nyc_geo.head()
```

```python
gp_poi['ll_round'] = gp_poi['geometry.location.lat'].round(3).astype(str) + ',' + gp_poi['geometry.location.lng'].round(3).astype(str)
gp_poi.head()
```

```python
# making sure all of our neighbourhood & borough information matches
num_rows = nyc_geo.shape[0]
row_count = 0

while row_count < num_rows:
    gp_poi.loc[gp_poi['Zip'] ==nyc_geo['Zipcode'][row_count], 'Neighbourhood'] = nyc_geo['Neighbourhood'][row_count]
    gp_poi.loc[gp_poi['ll_round'] ==nyc_geo['ll_round'][row_count], 'Neighbourhood'] = nyc_geo['Neighbourhood'][row_count]
    gp_poi.loc[gp_poi['Zip'] ==nyc_geo['Zipcode'][row_count], 'Borough'] = nyc_geo['Borough'][row_count]
    gp_poi.loc[gp_poi['ll_round'] ==nyc_geo['ll_round'][row_count], 'Borough'] = nyc_geo['Borough'][row_count]
    row_count +=1
```

```python
#Dropirrelevantorrepetitivecolumns
gp_poi.drop(columns=['business_status','icon','icon_background_color',
                    'icon_mask_base_uri','photos','place_id','reference','scope','types',
                    'geometry.viewport.northeast.lat','geometry.viewport.northeast.lng',
                    'geometry.viewport.southwest.lat','geometry.viewport.southwest.lng',
                    'opening_hours.open_now','plus_code.compound_code','plus_code.global_code',
                    'permanently_closed','result.formatted_address', 'll_round'],
                    inplace=True)
gp_poi.head()
```

```python
gp_poi.columns
```

```python
# Rename columns 
gp_poi.columns = ['Name','Price','Rating','Reviews','Address','Latitude','Longitude','Category','Borough','Zipcode','Neighborhood']

# Reorder columns 
gp_poi = gp_poi[['Name','Price','Rating','Reviews','Borough','Neighborhood','Address','Zipcode','Latitude','Longitude','Category']]
print(gp_poi.shape)
gp_poi.head()
```

```python
# Check for unique values

def unique_values(df):
    columns = df.columns
    for col in columns:
        print('Column Name:', col)
        print('# of Unique Values:', (len(df[col].unique())))
        print('Unique Values:', df[col].unique())
        print('==============================================================')
    
unique_values(gp_poi)
```

```python
# Check for nulls
gp_poi.isnull().sum()
```

There are null values in price, rating, reviews, borough, neighbourhood, address, and zipcode. However it has low impact, as it simply signifies that some points of interest have no price, rating, or review information, which does not affect our objective.

However, let's investigate the rows where location information is null, specifically boroughs.

```python
# Identify values where borough is null
not_borough = gp_poi[gp_poi['Borough'].isnull()]
not_borough['Address'].unique()
```

These locations are not actually considered part of the 5 boroughs: Bronx, Brooklyn, Manhattan, Queens and Staten Island, and so we will just drop them.

```python
# Fill null rows in Borough with borough from address DataFrame
gp_poi['Borough'].fillna('Drop', inplace=True)
```

```python
gp_poi.drop(gp_poi[gp_poi['Borough'] == 'Drop'].index, inplace=True)
```

```python
gp_poi.head()
```

```python
# Check for duplicates
len(gp_poi[gp_poi.duplicated(subset=['Name', 'Address'])].reset_index())
```

There are duplicates in the name value from the dataset, we will remove each additional instance of the area with the same name and street address.

```python
# # 5132 duplicate rows found, remove from DataFrame
gp_poi = gp_poi.drop_duplicates(subset=['Name', 'Address'], keep='last').reset_index(drop=True)

# Confirm removal of duplicates
len(gp_poi[gp_poi.duplicated()])
```
