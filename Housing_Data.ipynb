```python
# Data Cleaning
import pandas as pd
import numpy as np
```

```python
# Load datasets
med_house = pd.read_csv('clean_data/med_house_2021_c.csv') # median housing data
avg_house = pd.read_csv('clean_data/avg_house_2020_c.csv') # average housing data
nyc_geo = pd.read_csv('clean_data/nyc_geo_cleaned.csv') # nyc neighbourhood info
```

```python
med_house.drop(columns=['Unnamed: 0'], inplace=True)
med_house.head()
```

```python
avg_house.drop(columns=['Unnamed: 0'], inplace=True)
avg_house.head()
```

```python
"""
Loops through the nyc_geo (nyc boroughs) dataframe to look for matching neighbourhoods in the median_housing data.
Am also checking & cleaning average housing data at the same time
"""

num_rows = nyc_geo.shape[0]
row_count = 0

while row_count < num_rows:
    med_house.loc[med_house['Neighbourhood'] ==nyc_geo['Neighbourhood'][row_count], 'Borough'] = nyc_geo['Borough'][row_count]
    avg_house.loc[avg_house['Neighbourhood'] ==nyc_geo['Neighbourhood'][row_count], 'Borough'] = nyc_geo['Borough'][row_count]
    
    row_count +=1
```

```python
med_house = pd.read_csv('clean_data/med_housing_clean.csv') # median housing data
avg_house = pd.read_csv('clean_data/avg_housing_clean.csv') # average housing data
```

```python
med_house.head()
```

```python
avg_house.head()
```

```python
nyc_geo.head()
```

```python
# adding prefix for column names
med_house.columns=[f'Median_Cost_Per_{c}' for c in med_house]
avg_house.columns=[f'Avg_Cost_Per_{c}' for c in avg_house]
```

```python
nyc_geo = nyc_geo.merge(med_house, how='left', left_on='Neighbourhood', right_on='Median_Cost_Per_Neighbourhood')
```

```python
nyc_geo = nyc_geo.merge(avg_house, how='left', left_on='Neighbourhood', right_on='Avg_Cost_Per_Neighbourhood')
```

```python
nyc_geo.head()
```

```python
nyc_geo.info()
```

```python
nyc_geo.drop(columns=['Median_Cost_Per_Borough','Median_Cost_Per_Neighbourhood','Avg_Cost_Per_Borough','Avg_Cost_Per_Neighbourhood'],inplace=True)
```

```python
# there are a few duplicates, probably just from user error
nyc_geo[nyc_geo.duplicated(subset=['Neighbourhood'])]
```

```python
# grouping by mean to deal with duplicates, so we don't lose any data
nyc_means = nyc_geo.groupby(['Neighbourhood']).mean().reset_index()
```

```python
nyc_means.info()
```

```python
boroughs = nyc_geo[['Borough','Neighbourhood']]
```

```python
boroughs = boroughs.drop_duplicates()
```

```python
nyc_means
```

```python
nyc_geo = nyc_means.merge(boroughs, on='Neighbourhood', how='left')
```

```python
nyc_geo.info()
```

```python
nyc_geo.columns
```

```python
nyc_geo = nyc_geo[['Borough', 'Neighbourhood', 'Latitude', 'Longitude', 'Median_Cost_Per_Studio',
       'Median_Cost_Per_1_Bed', 'Median_Cost_Per_2_Bed', 'Avg_Cost_Per_Studio',
       'Avg_Cost_Per_1_Bed', 'Avg_Cost_Per_2_Bed', 'Avg_Cost_Per_3_Bed',]]
```
