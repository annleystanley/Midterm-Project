# Beginning Main Merge

### Joining POIs

#### Restaurants

```python
main_df = nb_ll_df.merge(resto_count, how='left', left_on='Neighbourhood', right_on='Restaurant_Count_Neighbourhood')
```

```python
main_df.head()
```

```python
main_df.drop(columns=['Restaurant_Count_Neighbourhood'], inplace=True)
```

```python
main_df = main_df.merge(resto_mean, how='left', left_on=['Neighbourhood'], right_on=['Restaurant_Mean_Neighbourhood'])
```

```python
main_df.drop(columns=['Restaurant_Mean_Neighbourhood'], inplace=True)
```

```python
main_df = main_df.merge(resto_sum, how='left', left_on=['Neighbourhood'], right_on=['Restaurant_Sum_Neighbourhood'])
```

```python
main_df.drop(columns=['Restaurant_Sum_Neighbourhood'], inplace=True)
```

#### Schools

```python
main_df = main_df.merge(school_count, how='left', left_on=['Neighbourhood'], right_on=['School_Count_Neighbourhood'])
main_df = main_df.merge(school_mean, how='left', left_on=['Neighbourhood'], right_on=['School_Mean_Neighbourhood'])
main_df = main_df.merge(school_sum, how='left', left_on=['Neighbourhood'], right_on=['School_Sum_Neighbourhood'])
```

```python
main_df.drop(columns=['School_Count_Neighbourhood', 'School_Mean_Neighbourhood', 'School_Sum_Neighbourhood'], inplace=True)
```

```python
main_df.head()
```

#### Parks

```python
main_df = main_df.merge(park_count, how='left', left_on=['Neighbourhood'], right_on=['Park_Count_Neighbourhood'])
main_df = main_df.merge(park_mean, how='left', left_on=['Neighbourhood'], right_on=['Park_Mean_Neighbourhood'])
main_df = main_df.merge(park_sum, how='left', left_on=['Neighbourhood'], right_on=['Park_Sum_Neighbourhood'])
```

```python
main_df.drop(columns=['Park_Count_Neighbourhood', 'Park_Mean_Neighbourhood', 'Park_Sum_Neighbourhood'], inplace=True)
```

#### Transit

```python
main_df = main_df.merge(transit_count, how='left', left_on=['Neighbourhood'], right_on=['Transit_Count_Neighbourhood'])
main_df = main_df.merge(transit_mean, how='left', left_on=['Neighbourhood'], right_on=['Transit_Mean_Neighbourhood'])
main_df = main_df.merge(transit_sum, how='left', left_on=['Neighbourhood'], right_on=['Transit_Sum_Neighbourhood'])
```

```python
main_df.drop(columns=['Transit_Count_Neighbourhood', 'Transit_Mean_Neighbourhood', 'Transit_Sum_Neighbourhood'], inplace=True)
```

```python
main_df.head()
```

#### Grocery

```python
main_df = main_df.merge(grocery_count, how='left', left_on=['Neighbourhood'], right_on=['Grocery_Count_Neighbourhood'])
main_df = main_df.merge(grocery_mean, how='left', left_on=['Neighbourhood'], right_on=['Grocery_Mean_Neighbourhood'])
main_df = main_df.merge(grocery_sum, how='left', left_on=['Neighbourhood'], right_on=['Grocery_Sum_Neighbourhood'])
```

```python
main_df.drop(columns=['Grocery_Count_Neighbourhood', 'Grocery_Mean_Neighbourhood', 'Grocery_Sum_Neighbourhood'], inplace=True)
```

```python
main_df.head()
```

### Joining Rats

```python
main_df = main_df.merge(rats_sum, how='left', left_on=['Neighbourhood'], right_on=['Rat_Sum_Neighbourhood'])
main_df.drop(columns=['Rat_Sum_Neighbourhood'], inplace=True)
main_df.head()
```

```python
main_df = main_df.merge(rats_count, how='left', left_on=['Neighbourhood'], right_on=['Rat_Count_Neighbourhood'])
main_df.drop(columns=['Rat_Count_Neighbourhood'], inplace=True)
main_df.head()
```

```python
main_df = main_df.merge(rats_mean, how='left', left_on=['Neighbourhood'], right_on=['Rat_Mean_Neighbourhood'])
main_df.drop(columns=['Rat_Mean_Neighbourhood'], inplace=True)
main_df.head()
```

```python
main_df = main_df.merge(uber_df_sum, how='left', left_on=['Neighbourhood'], right_on=['Uber_Sum_Neighbourhood'])
main_df.drop(columns=['Uber_Sum_Neighbourhood', 'Uber_Sum_Month','Uber_Sum_Day'], inplace=True)
main_df.head()
```

```python
main_df = main_df.merge(uber_df_count, how='left', left_on=['Neighbourhood'], right_on=['Uber_Count_Neighbourhood'])
main_df.drop(columns=['Uber_Count_Neighbourhood'], inplace=True)
main_df.head()
```

```python
main_df = main_df.merge(uber_df_mean, how='left', left_on=['Neighbourhood'], right_on=['Uber_Mean_Neighbourhood'])
main_df.drop(columns=['Uber_Mean_Neighbourhood', 'Uber_Mean_Ride Count'], inplace=True)
main_df.head()
```
